```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# Load an image using PIL
image_path = 'flowers/flowers/19_010.png'
original_image = Image.open(image_path)

# Display the original image using matplotlib
plt.imshow(original_image)
```

```python
from scipy.ndimage import rotate
def rotate_image_30_degrees(image):
    return rotate(image, angle=90)
```

### Display Original and Rotated Image

```python
rotated_image = rotate_image_30_degrees(original_image)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1) # 1 row, 2 columns, first subplot
plt.imshow(original_image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2) # 1 row, 2 columns, second subplot
plt.imshow(rotated_image, cmap='gray') # Use cmap='gray' for consistency if image is converted to grayscale during rotation
plt.title('Rotated Image (90 degrees)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
from PIL import ImageOps

mirrored_image = ImageOps.mirror(original_image)
flipped_image = ImageOps.flip(original_image)

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(original_image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(mirrored_image)
plt.title('Mirrored Image (Horizontal Flip)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(flipped_image)
plt.title('Flipped Image (Vertical Flip)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
original_width, original_height = original_image.size
zoom_factor = 1.2

new_width = int(original_width * zoom_factor)
new_height = int(original_height * zoom_factor)

zoomed_image = original_image.resize((new_width, new_height))

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(original_image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(zoomed_image)
plt.title(f'Zoomed Image ({zoom_factor}x)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
