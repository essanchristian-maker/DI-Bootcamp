```python
import pandas as pd
import zipfile
import requests
import io

def load_csv_from_zip_url(url):
    response = requests.get(url)
    z = zipfile.ZipFile(io.BytesIO(response.content))
    csv_name = [f for f in z.namelist() if f.endswith(".csv")][0]
    return pd.read_csv(z.open(csv_name))
```

```python
# Load Retail dataset
url_retail = "https://github.com/devtlv/Datasets-DA-Bootcamp-2-/raw/refs/heads/main/Week%204%20-%20Data%20Understanding/W4D3%20-%20Importing%20Data,%20Exporting%20D/Retail%20Data%20Analytics%20dataset.zip"

retail_df = load_csv_from_zip_url(url_retail)
retail_df.head()
```

```python
# Load Women's Clothing Reviews dataset
url_reviews = "https://github.com/devtlv/Datasets-DA-Bootcamp-2-/raw/refs/heads/main/Week%204%20-%20Data%20Understanding/W4D3%20-%20Importing%20Data,%20Exporting%20D/Women's%20E-Commerce%20Clothing%20Reviews.zip"

reviews_df = load_csv_from_zip_url(url_reviews)
reviews_df.head()
```

```python
print(retail_df.columns.tolist())
print(retail_df.head())
```

```python
# Sales trends — using MarkDown1 as main numerical indicator
print("=== TEMPERATURE TREND OVER TIME ===")
print(retail_df.groupby("Date")["Temperature"].mean().sort_index().head(10))

# Store performance — by average CPI and Unemployment
print("\n=== STORE PERFORMANCE (Avg Unemployment) ===")
print(retail_df.groupby("Store")["Unemployment"].mean().sort_values())

# Holiday purchasing habits
print("\n=== AVERAGE TEMPERATURE BY HOLIDAY ===")
print(retail_df.groupby("IsHoliday")["Temperature"].mean())

print("\n=== AVERAGE FUEL PRICE BY HOLIDAY ===")
print(retail_df.groupby("IsHoliday")["Fuel_Price"].mean())
```


### Structured Data : Retail Dataset
Columns : Store, Date, Temperature, Fuel_Price, MarkDown1-5, CPI, Unemployment, IsHoliday

- **Trend analysis** : Temperature and Fuel_Price tracked over time by Date.
- **Store performance** : Unemployment rate compared across stores.
- **Customer habits** : Holiday vs non-holiday periods compared via
  Temperature and Fuel_Price averages.
- This data is clean, numerical and easy to aggregate.

### Challenges
- No direct Sales column — insights are indirect (economic indicators).
- MarkDown columns contain many NaN values, limiting their use.

```python
def load_csv_latin(url):
    response = requests.get(url)
    z = zipfile.ZipFile(io.BytesIO(response.content))
    csv_name = [f for f in z.namelist() if f.endswith(".csv")][0]
    return pd.read_csv(z.open(csv_name), encoding="latin-1")

url_ecommerce = "https://github.com/devtlv/Datasets-DA-Bootcamp-2-/raw/refs/heads/main/Week%204%20-%20Data%20Understanding/W4D3%20-%20Importing%20Data,%20Exporting%20D/E-Commerce%20Dataset.zip"

ecommerce_df = load_csv_latin(url_ecommerce)

print("=== FIRST ROWS ===")
print(ecommerce_df.head())

print("\n=== BASIC INFO ===")
print(f"Rows    : {ecommerce_df.shape[0]}")
print(f"Columns : {ecommerce_df.shape[1]}")
print(f"Column names : {ecommerce_df.columns.tolist()}")
print(ecommerce_df.info())
```

```python
structured = ecommerce_df.select_dtypes(include=['float64', 'int64', 'datetime64']).columns.tolist()
categorical = ecommerce_df.select_dtypes(include=['object']).columns.tolist()

print("Structured (numerical/date) columns :", structured)
print("Categorical columns                 :", categorical)
```


### Structured Data Columns
| Column | Type | Reasoning |
|---|---|---|
| Quantity | Numerical | Number of units purchased per transaction |
| UnitPrice | Numerical | Price per unit in £ |
| CustomerID | Numerical | Unique customer identifier |
| InvoiceDate | Date | Should be converted to datetime for trend analysis |

### Categorical Columns (also structured)
| Column | Type | Reasoning |
|---|---|---|
| InvoiceNo | Categorical | Unique transaction identifier |
| StockCode | Categorical | Unique product code |
| Description | Categorical | Product name — borders on unstructured |
| Country | Categorical | Customer's country |

### Suggested Unstructured Data to Complement this Dataset
| Unstructured Data | How it complements structured data |
|---|---|
| Customer reviews | Reveal satisfaction beyond Quantity and UnitPrice |
| Product descriptions | Enrich StockCode and Description for better categorization |
| Customer support chats | Identify issues linked to specific InvoiceNo or StockCode |
| Social media comments | Capture sentiment trends by Country |

### How Unstructured Data Would Be Used
- **Reviews + Quantity** : high sales volume but negative reviews
  signals a quality problem.
- **Product descriptions + StockCode** : auto-classify products
  and improve recommendations.
- **Support chats + Country** : detect regional issues
  and adapt customer service accordingly.

Combining both types gives a 360° view of customer behavior
that structured data alone cannot provide.

- **holiday** is semi-structured : it is either None or a named
  string — it could be encoded as a binary column (0/1).
- **weather_main** is semi-structured : it looks like a fixed
  category but is stored as free text without strict encoding.

```python
# Step 1 : Install and import Faker
!pip install faker

from faker import Faker
import pandas as pd
import random

fake = Faker()

# Step 2 : Generate 500 products
products = []

for i in range(1, 501):
    products.append({
        "ProductID"   : i,
        "Name"        : fake.catch_phrase(),
        "Description" : fake.text(max_nb_chars=100),
        "Price"       : round(random.uniform(5.99, 999.99), 2)
    })

# Step 3 : Create DataFrame
products_df = pd.DataFrame(products)

print(f"Shape : {products_df.shape}")
products_df.head()
```

 **Faker** generates realistic fake text data (names, descriptions).
- **random.uniform(5.99, 999.99)** produces realistic price ranges.
- **ProductID** is a unique sequential integer from 1 to 500.
- The final DataFrame has 500 rows and 4 columns :
  ProductID, Name, Description, Price.
