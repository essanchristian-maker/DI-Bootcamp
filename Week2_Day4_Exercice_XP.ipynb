# Exercise 1: Understanding Data Visualization

## Why is data visualization important in data analysis?

Data visualization transforms raw numbers into visual forms that the human
brain can interpret far more quickly than tables of figures. It plays
several key roles in data analysis:

- **Pattern detection**: trends, cycles, or seasonality that are nearly
  invisible in a table of thousands of rows often become obvious on a chart.
- **Anomaly detection**: outliers stand out immediately on a boxplot or
  scatter plot, whereas they can be easily missed in summary statistics.
- **Effective communication**: a non-technical decision-maker can grasp a
  well-designed chart in seconds, while a table of statistics requires time
  and expertise to interpret.
- **Guiding exploration (EDA)**: visualizing data early helps formulate the
  right hypotheses to test, even before running any formal statistical test.
- **Validating model assumptions**: for example, a scatter plot before
  fitting a linear regression shows whether the relationship is actually
  linear, or whether a different model would be more appropriate.

> **Example**: in the airplane crash analysis, a skewness value of 4.61 is
> an abstract number — but a histogram with mean and median lines overlaid
> makes that asymmetry immediately understandable, even without knowing the
> term "skewness."

In short, visualization is not just a tool for presenting final results —
it is also a discovery tool used throughout the analysis process itself.

## What is the purpose of a line graph in data visualization?

A **line graph** connects data points with line segments, typically in a
sequential order (most often time). Its main purpose is to show how a
continuous variable **evolves over time** (or another ordered variable),
making it possible to reveal:

- **Trends** (increasing, decreasing, or stable)
- **Cycles or seasonality** (e.g., sales spiking every December)
- **Turning points** (sudden changes in behavior)
- **Comparisons between groups**, when multiple lines are plotted together

> **Example**: a line graph of "Number of crashes per year (1908–2023)"
> makes the rise through the mid-20th century and the sharp decline after
> 2000 immediately visible — a pattern that would be far harder to spot in
> a table of 115 yearly values.

### When to use a line graph (and when not to)

| Use a line graph when... | Avoid a line graph when... |
|---|---|
| Data is ordered (time, age, etc.) | Categories are independent/unordered (e.g., countries, colors) → prefer a bar chart |
| You want to show a continuous trend | You're comparing totals or proportions at a single point → prefer a bar or pie chart |
| Few groups need to be compared at once | Many groups need to be compared simultaneously (the chart becomes unreadable) |

```python
# ============================================================
# EXERCICE 2 : Line Plot - Variation de température sur une semaine
# ============================================================

import matplotlib.pyplot as plt

# ------------------------------------------------------------
# 1) Données : températures pour chaque jour de la semaine
# ------------------------------------------------------------
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperatures = [72, 74, 76, 80, 82, 78, 75]

# ------------------------------------------------------------
# 2) Création du graphique
# ------------------------------------------------------------
plt.figure(figsize=(8, 5))

# marker='o' ajoute un point visible à chaque jour (lisibilité)
# linewidth contrôle l'épaisseur de la ligne, color sa couleur
plt.plot(days, temperatures, marker='o', linewidth=2, color='darkorange')

# ------------------------------------------------------------
# 3) Labels des axes (obligatoire pour qu'un graphique soit lisible)
# ------------------------------------------------------------
plt.xlabel("Day")
plt.ylabel("Temperature (°F)")

# ------------------------------------------------------------
# 4) Titre du graphique
# ------------------------------------------------------------
plt.title("Temperature Variation Over a Week")

# ------------------------------------------------------------
# 5) Affichage
# ------------------------------------------------------------
plt.grid(True, alpha=0.3)  # grille légère : aide à lire les valeurs précisément
plt.show()
```

```python
# ============================================================
# EXERCICE 3 : Bar Chart - Ventes mensuelles d'un magasin
# ============================================================

import matplotlib.pyplot as plt

# ------------------------------------------------------------
# 1) Données : ventes pour chaque mois
# ------------------------------------------------------------
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [5000, 5500, 6200, 7000, 7500]

# ------------------------------------------------------------
# 2) Création du bar chart
# ------------------------------------------------------------
plt.figure(figsize=(8, 5))

# plt.bar() trace des barres verticales : une par catégorie (mois)
plt.bar(months, sales, color='steelblue', edgecolor='black')

# ------------------------------------------------------------
# 3) Labels des axes
# ------------------------------------------------------------
plt.xlabel("Month")
plt.ylabel("Sales Amount ($)")

# ------------------------------------------------------------
# 4) Titre du graphique
# ------------------------------------------------------------
plt.title("Monthly Sales for Retail Store")

# ------------------------------------------------------------
# 5) Affichage
# ------------------------------------------------------------
plt.grid(axis='y', alpha=0.3)  # grille uniquement horizontale : aide à lire les hauteurs de barres
plt.show()
```

```python
# ============================================================
# Étape 0 : Téléchargement et inspection du dataset
# ============================================================
# Ce script télécharge le fichier ZIP depuis GitHub, l'extrait,
# charge le CSV dans un DataFrame, puis affiche les colonnes
# et les premières lignes pour vérifier le format exact.

import requests   # pour télécharger le fichier depuis l'URL
import zipfile    # pour extraire le contenu du ZIP
import io         # pour manipuler le ZIP directement en mémoire
import pandas as pd

# --- URL du fichier ZIP sur GitHub ---
url = "https://github.com/devtlv/Datasets-GEN-AI-Bootcamp/raw/refs/heads/main/Week%203/W3D1%20-%20Data%20Visualization/Student%20Mental%20health.zip"

# --- Téléchargement du fichier ---
response = requests.get(url)
response.raise_for_status()  # déclenche une erreur si le téléchargement échoue

# --- Extraction du ZIP en mémoire (sans créer de fichier intermédiaire) ---
zip_file = zipfile.ZipFile(io.BytesIO(response.content))

# Affiche la liste des fichiers contenus dans le ZIP
print("Fichiers trouvés dans le ZIP :")
print(zip_file.namelist())

# --- Chargement automatique du premier fichier CSV trouvé ---
csv_filename = [name for name in zip_file.namelist() if name.endswith(".csv")][0]
df = pd.read_csv(zip_file.open(csv_filename))

# --- Inspection du DataFrame ---
print("\nColonnes du dataset :")
print(df.columns.tolist())

print("\nAperçu des premières lignes :")
print(df.head())
```

```python
# ============================================================
# Exercice 5 : Comparaison du niveau d'anxiété selon le genre
# ============================================================

# --- 1. Importation des bibliothèques nécessaires ---
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- 2. Préparation des données ---
# La colonne "Do you have Anxiety?" contient du texte ("Yes"/"No").
# On la convertit en valeur numérique (1 = Oui, 0 = Non) pour pouvoir
# calculer une proportion (moyenne) par genre.
df["Anxiety_numeric"] = df["Do you have Anxiety?"].map({"Yes": 1, "No": 0})

# --- 3. Création du bar plot avec Seaborn ---
# Par défaut, sns.barplot() calcule automatiquement la moyenne de "y"
# pour chaque catégorie de "x" -> ici, ça donne directement la
# proportion d'étudiants en anxiété, par genre.
plt.figure(figsize=(7, 5))

sns.barplot(
    data=df,
    x="Choose your gender",     # catégorie sur l'axe X (genre)
    y="Anxiety_numeric",         # proportion d'anxiété sur l'axe Y
    palette="Set2",              # palette de couleurs personnalisée
    errorbar=None                # masque les barres d'erreur (plus lisible)
)

# --- 4. Personnalisation du graphique ---
plt.title("Proportion d'étudiants souffrant d'anxiété par genre", fontsize=14)
plt.xlabel("Genre")
plt.ylabel("Proportion ayant de l'anxiété")
plt.ylim(0, 1)  # car c'est une proportion, entre 0 et 1
plt.tight_layout()

# --- 5. Affichage du graphique ---
plt.show()

# Note : si "errorbar" génère une erreur (ancienne version de Seaborn),
# remplacer ce paramètre par : ci=None
```

```python
# ============================================================
# Exercice 6 : Relation entre l'âge et les crises de panique
# ============================================================

# --- 1. Importation des bibliothèques nécessaires ---
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- 2. Conversion des réponses en valeurs numériques ---
# La colonne s'appelle "Do you have Panic attack?" dans le dataset.
# On convertit "Yes"/"No" en 1/0 pour pouvoir les placer sur l'axe Y.
df["Panic_attack_numeric"] = df["Do you have Panic attack?"].map({"Yes": 1, "No": 0})

# --- 3. Création du scatter plot avec Seaborn ---
plt.figure(figsize=(8, 5))

sns.scatterplot(
    data=df,
    x="Age",                       # âge des étudiants sur l'axe X
    y="Panic_attack_numeric",      # 0 ou 1 sur l'axe Y
    alpha=0.4,                     # transparence : utile car beaucoup
                                    # de points se superposent (0 ou 1)
    s=80,                           # taille des points
    color="darkorange"
)

# --- 4. Personnalisation pour améliorer la lisibilité ---
plt.title("Relation entre l'âge et les crises de panique", fontsize=14)
plt.xlabel("Âge")
plt.ylabel("Crise de panique (0 = Non, 1 = Oui)")
plt.yticks([0, 1], ["Non", "Oui"])  # remplace 0/1 par des étiquettes claires
plt.tight_layout()

# --- 5. Affichage du graphique ---
plt.show()

# Remarque : comme la variable Y n'a que deux valeurs possibles (0 ou 1),
# le scatter plot classique montre des points superposés à chaque âge.
# Si tu veux mieux voir la densité, tu peux ajouter un léger "jitter" :
# sns.stripplot(data=df, x="Age", y="Panic_attack_numeric", jitter=0.05)
```
