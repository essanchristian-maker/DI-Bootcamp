Daily Challenge : NumPy, Pandas, and M


1. Data Preparation

```python
import numpy as np
import pandas as pd

# Configuration pour la reproductibilité
np.random.seed(42)

# 1. Générer les données avec NumPy (10 villes, 12 mois)
# Les températures s'échelonnent de -5 à 35 degrés Celsius
data = np.random.uniform(low=-5, high=35, size=(10, 12))


# 2. Définir les noms des villes et des mois
# Correction : Ajout de la virgule entre SASSANDRA et DALOA
villes = ["ABIDJAN", "BOUAKE", "MAN", "ODIENNE", "BONOUA", "BINGEVILLE", "KONG", "SASSANDRA", "DALOA", "BONGOUANOU"]
mois = ["Janv", "Févr", "Mars", "Avr", "Mai", "Juin", "Juil", "Août", "Sept", "Oct", "Nov", "Déc"]

# 3. Créer le DataFrame Pandas
weather_df = pd.DataFrame(np.round(data,0 ), index=villes, columns=mois)

# Affichage du résultat
display(weather_df)
```

2. Data Analysis:

```python
x = weather_df.mean(1)
print(np.round(x,2))
```

```python
Ville = x.idxmax()
temperature  = x.max()
Temperature_moyenne = x.idxmax()

print(f"The city with the highest average temperature is {Ville} with {np.round(temperature,2) } " )



```

```python
Ville_b = x.idxmin()
temperature_b  = x.min()
Temperature_moyenne = x.idxmin()
print(f"The city with the lowest average temperature is {Ville_b} with {np.round(temperature_b,2) } " )

```

3. Visualisation des données :


```python
import matplotlib.pyplot as plt

# Inversion (transposition) du DataFrame pour avoir les mois sur l'axe X
weather_df_visu = weather_df.T

# Configuration de la taille de la figure
plt.figure(figsize=(11, 6))

# Palette de couleurs (nuances de vert et sarcelle comme sur la capture)
couleurs_ci = [
    "#4db6ac", "#00897b", "#004d40", "#80cbc4", "#26a69a",
    "#00695c", "#009688", "#4db6ac", "#00796b", "#00b0ff"
]

# Tracer chaque courbe de ville
for i, ville in enumerate(weather_df_visu.columns):
    # Style spécifique en pointillés pour KONG (index 6) et BONGOUANOU (index 9)
    if ville in ["KONG", "BONGOUANOU"]:
        plt.plot(weather_df_visu.index, weather_df_visu[ville], label=ville,
                 color="#009688", linewidth=2.5, linestyle="--")
    else:
        # Les autres villes ont des lignes pleines et plus fines
        plt.plot(weather_df_visu.index, weather_df_visu[ville], label=ville,
                 color=couleurs_ci[i % len(couleurs_ci)], linewidth=1.3)

# --- Personnalisation esthétique (Fidèle à la capture d'écran) ---

# Titres et étiquettes
plt.title("Monthly Temperature Trends for Each City", fontsize=14, pad=15, loc="center")
plt.xlabel("Month", fontsize=11, labelpad=8)
plt.ylabel("Temperature (°C)", fontsize=11, labelpad=8)

# Grille d'arrière-plan discrète en pointillés
plt.grid(True, linestyle="--", alpha=0.4, color="#cccccc")

# Ajustement de l'échelle de l'axe Y (de -5 à 35 avec des pas de 5)
plt.ylim(-7, 36)
plt.yticks(range(-5, 36, 5))

# Suppression des bordures du haut et de droite (Style Minimaliste / Seaborn)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#333333')
ax.spines['bottom'].set_color('#333333')

# Emplacement de la légende (En bas à gauche)
plt.legend(loc="lower left", fontsize=9, framealpha=0.8)

# Optimisation de l'espace et affichage
plt.tight_layout()
plt.show()
```

Summary Report of Observations

Coldest City: Conversely, BOUAKE records the lowest annual average at around 9.5°C.

2. Key Trends and Observations
"Spaghetti Plot" Effect: The graph highlights a highly dense crisscrossing of lines. This demonstrates strong instability and significant month-to-month variability within the same city, which is typical of purely randomly generated data (np.random.uniform).

Artificial Seasonality: Contrary to actual Ivorian climate patterns (which exhibit milder real-world temperature variations), the model shows sharp drops

Value of the Color Palette: Assigning a unique color to each locality and adding markers helps break through the visual clutter, allowing users to precisely track the temperature peaks of each individual city.
