## 1. What is Machine Learning?

Machine Learning (ML) is a branch of artificial intelligence that enables computer systems to **learn automatically from data**, without being explicitly programmed for each task. Rather than following rigid rules, an ML model identifies statistical patterns in data and improves its performance through experience.

> 💡 **Key Definition**: "Machine Learning gives computers the ability to learn without being explicitly programmed." — *Arthur Samuel, 1959*

### Why is ML Essential for Data Analysts?

- **Automation** of repetitive analyses on large volumes of data
- **Pattern detection** invisible to the human eye in complex datasets
- **Quantitative predictions**: anticipating behaviors, trends, and risks
- **Data-driven decision making**: replacing intuition with statistical evidence
- **Scalability**: a model trained once can process millions of new cases

## 3. The 3 Types of Machine Learning

| Criterion | 🟢 Supervised | 🟠 Unsupervised | 🟣 Reinforcement |
|-----------|--------------|----------------|-----------------|
| **Data** | Labeled (with labels) | Unlabeled | Rewards / Penalties |
| **Goal** | Predict a known output | Discover hidden structures | Maximize cumulative reward |
| **Example** | Cancer detection (0/1) | Customer segmentation | Robot learning to walk |
| **Typical Algo** | Log. Reg., SVM, Forest | K-Means, PCA, DBSCAN | Q-Learning, PPO |

### 3.1 Supervised Learning

> 📌 The model learns on **labeled data**: each training example is paired with a known answer (label). The algorithm learns to reproduce the input → output mapping.

- **Classification**: predict a category (spam/not-spam, sick/healthy, fraud/legitimate)
- **Regression**: predict a continuous value (house price, temperature, stock price)
- **Common algorithms**: Logistic Regression, SVM, Random Forest, Neural Networks

**Example:** A model trained on 10,000 manually labeled emails (spam / not-spam) learns to recognize keywords and patterns associated with each category. It can then automatically classify new incoming emails.

### 3.2 Unsupervised Learning

> 📌 The model works on data **without labels**. It seeks to discover hidden structures, natural groups (clusters), or compact representations within the data.

- **Clustering**: group similar data points together (K-Means, DBSCAN, HAC)
- **Dimensionality reduction**: compress information (PCA, t-SNE, UMAP)
- **Anomaly detection**: identify atypical points without a reference example

**Example:** A retailer groups its customers into segments (loyal, occasional, deal-hunters) based on their purchasing habits, without having predefined those segments beforehand.

### 3.3 Reinforcement Learning

> 📌 An **agent** makes decisions in an environment and receives rewards or penalties based on its actions. It learns through trial-and-error to maximize cumulative reward.

- The agent observes a state → chooses an action → receives a reward → updates its strategy
- **Common algorithms**: Q-Learning, Deep Q-Network (DQN), PPO, AlphaZero
- **Applications**: video games, robotics, algorithmic trading, energy management

**Example:** DeepMind's AlphaGo learned to play the game of Go by playing millions of games against itself, receiving a positive reward for each win. It ultimately surpassed the world's best human players.

## 4. The ML Model Development Process

```
① Data Collection
   └─ Gather raw data from various sources (CSV, DB, API, sensors...)
          │
          ▼
② Feature Selection
   └─ Identify relevant variables, remove redundancy, feature engineering
          │
          ▼
③ Data Preprocessing
   └─ Handle missing values, encode categorical variables, normalize/standardize
          │
          ▼
④ Model Selection
   └─ Choose the right algorithm for the problem type and data available
          │
          ▼
⑤ Model Training
   └─ Fit the model on the training set to minimize prediction error
          │
          ▼
⑥ Model Evaluation
   └─ Measure performance on the test set (Accuracy, F1, RMSE...)
          │
          ▼
⑦ Deployment & Monitoring
   └─ Put the model into production and track data drift over time
```

### Focus on the 3 Key Stages

#### ② Feature Selection

Feature selection involves identifying the most relevant input variables for predicting the target. Redundant or low-quality features degrade model performance and increase computation time.

- **Filter methods**: Pearson correlation, Chi-square test, mutual information
- **Wrapper methods**: sequential search (forward / backward selection)
- **Embedded methods**: LASSO (L1 regularization), feature importance in decision trees
- **Feature Engineering**: creating new variables from existing ones

> ⚠️ **Golden Rule:** *Garbage In → Garbage Out* — data and feature quality determines model quality. Good feature engineering is often more valuable than a more complex algorithm.

#### ④ Model Selection

The choice of algorithm depends on the problem type, dataset size, desired interpretability, and performance constraints. There is no universally superior algorithm (**No Free Lunch Theorem**).

- **Classification** → Logistic Regression, SVM, Random Forest, XGBoost
- **Regression** → Linear Regression, Ridge, LASSO, Gradient Boosting
- **Text data** → Naive Bayes, Transformers (BERT, GPT)
- **Cross-validation** (k-fold): robustly compare multiple models
- **Hyperparameter tuning**: GridSearchCV, RandomizedSearchCV, Optuna

#### ⑥ Model Evaluation

Evaluation measures the model's ability to generalize to unseen data (test set). Each metric captures a different aspect of performance:

| Metric | Formula | When to Use |
|--------|---------|-------------|
| **Accuracy** | (TP+TN) / Total | Balanced classes — overall view |
| **Precision** | TP / (TP+FP) | Minimize false alarms (spam) |
| **Recall** | TP / (TP+FN) | Minimize missed cases (cancer) |
| **F1-Score** | 2×P×R / (P+R) | Balance between precision and recall |
| **RMSE / MAE** | √(Σ(y-ŷ)²/n) | Regression problems |
| **AUC-ROC** | Area under the ROC curve | Evaluate classifier ranking ability |

## 5. Conclusion

### ✅ Key Takeaways

1. ML enables systems to learn patterns from data automatically.
2. Three families: **Supervised** (labeled data), **Unsupervised** (no labels), **Reinforcement** (rewards).
3. The ML process is iterative: data → features → model → evaluation → deployment.
4. Data and feature quality is more critical than the choice of algorithm.
5. Evaluation must always be performed on data **not seen** during training.
