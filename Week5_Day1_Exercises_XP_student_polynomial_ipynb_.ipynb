## 📝 Pourquoi le Deep Learning excelle sur les données non structurées

Le **deep learning** domine le traitement des données non structurées (images, audio, texte, vidéo) principalement grâce au principe de **representation learning** : au lieu de demander à un expert humain de définir manuellement les caractéristiques pertinentes, le réseau apprend lui-même quelles représentations sont utiles pour la tâche donnée. Cette capacité s'appuie sur l'extraction de **hierarchical features** — chaque couche successive combine les sorties de la précédente pour construire des concepts de plus en plus abstraits. Par exemple, dans un CNN appliqué à des images, les premières couches détectent des bords et des gradients de couleur, les couches intermédiaires reconnaissent des formes et des textures, tandis que les couches profondes identifient des objets complets comme un visage ou un véhicule. Cette hiérarchie est impossible à concevoir manuellement avec des techniques traditionnelles de feature engineering, car le nombre de combinaisons pertinentes est astronomique. De plus, le partage de poids (convolutions, mécanismes d'attention) permet aux réseaux neuronaux de **généraliser efficacement** à travers les variations naturelles des données (translation, rotation, accent, synonymes), ce que les modèles ML classiques peinent à capturer sans un pré-traitement massif et coûteux.

## 🌟 Exercise 3: Creating the Dataset and Visualizing the Data

**As stated in the exercise**  
Objective. Generate a dataset with noise and visualize it.  
Instructions. Import numpy, matplotlib.pyplot, and mean_squared_error from sklearn.metrics. Create 20 points using \(y = -x^2\) with Gaussian noise \(\mathcal{N}(0, 0.05)\). Plot the points with a scatter plot. Split into a training set with the first 12 points and a test set with the last 8 points.


```python
# PREFILLED: just execute
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

np.random.seed(0)
x = np.arange(-1, 1, 0.1)  # 20 points from -1 to <1 step 0.1
print("x shape:", x.shape)
```

```python
# --- Génération de y avec bruit ---
# La formule théorique est y = -x², une parabole inversée.
# On ajoute un bruit gaussien pour simuler des données réelles imparfaites.
# np.random.normal(0, 0.05, len(x)) :
#   - moyenne = 0    -> pas de décalage systématique
#   - écart-type = 0.05 -> dispersion du bruit autour de la courbe
#   - len(x) = 20    -> une valeur de bruit par point x
y = -x**2 + np.random.normal(0, 0.05, len(x))

print("y shape:", y.shape)

# --- Validation rapide ---
# On vérifie que y a bien la même forme que x (20 points) et que tout est cohérent.
print(f"Premières valeurs de y : {np.round(y[:5], 4)}")  # 4 décimales pour lisibilité
print(f"x a {len(x)} points, y a {len(y)} points -> correspondance : {len(x) == len(y)}")
```

```python
# PREFILLED: just execute (scatter plot of the dataset)
plt.figure(figsize=(5,4))
plt.scatter(x, y)
plt.title("Noisy dataset sampled from y = -x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.tight_layout()
plt.show()
```

```python
# --- Séparation en jeu d'entraînement et jeu de test ---
# Pourquoi faire cette séparation ?
# - Train set : sert à entraîner (ajuster) le modèle.
# - Test set  : sert à évaluer le modèle sur des données qu'il n'a jamais vues.
#   Cela permet de vérifier si le modèle généralise bien.

# Consigne : 12 premiers points pour l'entraînement, 8 derniers pour le test.
# x et y ont 20 éléments, indexés de 0 à 19.
# x[:12]  -> indices 0 à 11 (12 premiers points)
# x[12:]  -> indices 12 à 19 (8 derniers points)
x_train, y_train = x[:12], y[:12]
x_test,  y_test  = x[12:], y[12:]

print("train sizes:", x_train.shape, y_train.shape, "test sizes:", x_test.shape, y_test.shape)

# --- Validation ---
# On vérifie que la somme des tailles correspond bien au total (20 points)
# et que les deux ensembles ne se chevauchent pas.
print(f"Nombre total de points : {len(x)}")
print(f"Train : {len(x_train)} points | Test : {len(x_test)} points")
print(f"Train + Test = {len(x_train) + len(x_test)} (doit être 20)")
print(f"Premier x_test : {x_test[0]:.1f} (doit être 0.2, juste après le dernier x_train)")
print(f"Dernier x_train : {x_train[-1]:.1f} (doit être 0.1)")
```

**Learning point**  
Visualizing noisy data helps separate signal from noise. Splitting into train and test allows you to detect when a model overfits noise instead of learning the underlying pattern.


## 🌟 Exercise 4: Fitting Polynomial Models of Different Degrees

**As stated in the exercise**  
Objective: Fit polynomial models of varying degrees and observe overfitting.  

Instructions: Define `polynomial_fit(degree)` that returns the fitted polynomial coefficients from training data. Define `plot_polyfit(degree)` that draws the training set, the test set, and the fitted polynomial curve. Visualize fits for degrees 1, 7, and 11. Observe behavior on training and test sets as degree increases.


```python
# --- Implémentation de polynomial_fit ---
# Prend le degré du polynôme souhaité.
# Retourne une fonction polynomiale ajustée sur les données d'entraînement.
def polynomial_fit(degree: int):
    # np.polyfit calcule les coefficients du polynôme par moindres carrés.
    coeffs = np.polyfit(x_train, y_train, degree)

    # np.poly1d transforme le tableau de coefficients en fonction appelable.
    # Exemple : np.poly1d([1, 0, -1]) crée f(x) = x² - 1
    return np.poly1d(coeffs)

# --- Validation rapide ---
# Test avec degré 1 (une simple droite).
poly_test = polynomial_fit(1)
print(f"Coefficients du polynôme : {poly_test}")
print(f"Type retourné : {type(poly_test).__name__}")  # Doit être poly1d
print(f"poly_test(0) = {poly_test(0):.4f}")
```

```python
# --- Implémentation de plot_polyfit ---
# Prend le degré du polynôme.
# Affiche un graphique avec :
#   - points d'entraînement (bleu)
#   - points de test (rouge)
#   - courbe polynomiale ajustée (ligne)
def plot_polyfit(degree: int):
    # 1) Ajuster le polynôme sur les données d'entraînement
    poly = polynomial_fit(degree)

    # 2) Créer un linspace dense pour une courbe lisse
    # On couvre tout l'intervalle [-1, 1] avec 200 points.
    x_curve = np.linspace(-1, 1, 200)

    # 3) Évaluer le polynôme sur ces points
    y_curve = poly(x_curve)

    # 4) Tracer : points d'entraînement, points de test, courbe
    plt.figure(figsize=(6, 4))
    plt.scatter(x_train, y_train, color='blue', label='Train', zorder=3)
    plt.scatter(x_test, y_test, color='red', label='Test', zorder=3)
    plt.plot(x_curve, y_curve, color='green', linewidth=2, label=f'Degree {degree}', zorder=2)

    # 5) Affichage avec titre et légende
    plt.title(f"Polynomial fit - Degree {degree}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
# --- Visualisation des trois degrés demandés ---
# Degré 1 : une simple droite, sous-apprentissage probable.
# Degré 7 : polynôme modéré, devrait bien suivre la parabole.
# Degré 11 : polynôme élevé, risque de surapprentissage.
plot_polyfit(1)
plot_polyfit(7)
plot_polyfit(11)
```

**Learning point**  
Higher degree polynomials can perfectly fit training data yet fail to generalize. This is overfitting. You should always check test error, not just training error.

## 🌟 Exercise 5: Cross-Validation to Find the Optimal Degree

**As stated in the exercise**  
Objective: Use cross-validation to determine the best polynomial degree.  

Instructions: Loop degrees from 1 to 11. For each degree, compute the RMSE on the training and test sets. Store results in a list. Plot RMSE vs degree with a logarithmic y axis. Identify the degree minimizing test RMSE and confirm it matches the true model \(y=-x^2\).


```python
# --- Définition de la fonction RMSE ---
# Prend les vraies valeurs y_true et les prédictions y_pred.
# Retourne la racine de l'erreur quadratique moyenne.
def rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred)**2))

# --- Validation rapide ---
# Test avec des valeurs simples pour vérifier le calcul.
test_vrai = np.array([0, 0, 0])
test_pred = np.array([1, 2, 3])
# Erreurs : 1, 2, 3 -> carrés : 1, 4, 9 -> moyenne : 14/3 ≈ 4.67 -> racine ≈ 2.16
print(f"RMSE test : {rmse(test_vrai, test_pred):.4f} (attendu ≈ 2.1602)")
```

```python
# Initialize a list to store results (degree, train_rmse, test_rmse)
rows = []

# Loop degrees from 1 to 11
for degree in range(1, 12):
    # Fit the polynomial model
    poly = polynomial_fit(degree)

    # Predict on training and test sets
    y_train_pred = poly(x_train)
    y_test_pred = poly(x_test)

    # Calculate RMSE
    train_rmse = rmse(y_train, y_train_pred)
    test_rmse = rmse(y_test, y_test_pred)

    # Store results
    rows.append([degree, train_rmse, test_rmse])

# Extract data for plotting
degs = [r[0] for r in rows]
rmse_tr = [r[1] for r in rows]
rmse_te = [r[2] for r in rows]

# Plotting RMSE curves
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.plot(degs, rmse_tr, label="train RMSE")
plt.plot(degs, rmse_te, label="test RMSE")
plt.yscale("log")
plt.xlabel("polynomial degree")
plt.ylabel("RMSE")
plt.title("RMSE vs degree")
plt.legend()
plt.tight_layout()
plt.show()

# Find the degree with the minimum test RMSE
min_test_rmse_degree = degs[np.argmin(rmse_te)]
print(f"The degree with the minimum test RMSE is: {min_test_rmse_degree}")
```

**To-Do:** Report the degree with the minimum test RMSE and explain in 2–3 sentences how this connects to the true model \(y=-x^2\) and the bias–variance trade off.


```python
# --- Calcul du RMSE pour chaque degré de 1 à 11 ---
degrees = range(1, 12)  # 1, 2, ..., 11
rows = []                # Chaque élément : (degré, rmse_train, rmse_test)

for d in degrees:
    # Ajuster le polynôme de degré d sur les données d'entraînement
    poly = polynomial_fit(d)

    # Prédire sur l'ensemble d'entraînement et calculer le RMSE
    y_train_pred = poly(x_train)
    rmse_train = rmse(y_train, y_train_pred)

    # Prédire sur l'ensemble de test et calculer le RMSE
    y_test_pred = poly(x_test)
    rmse_test = rmse(y_test, y_test_pred)

    # Stocker les résultats
    rows.append((d, rmse_train, rmse_test))

# --- Affichage des résultats pour vérification ---
print("Degré | RMSE Train | RMSE Test")
print("-" * 35)
for d, rmse_tr, rmse_te in rows:
    print(f"  {d:2d}  |   {rmse_tr:.4f}   |   {rmse_te:.4f}")

# --- Trouver le degré qui minimise le RMSE test ---
best_degree = min(rows, key=lambda r: r[2])  # r[2] = rmse_test
print(f"\nDegré avec le plus petit RMSE test : {best_degree[0]} (RMSE = {best_degree[2]:.4f})")
```
