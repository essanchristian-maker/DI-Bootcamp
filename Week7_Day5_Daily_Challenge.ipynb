# Precision Prompting Challenge

## Step 1: Prompt Controlling Tone, Format, and Length

**Prompt:**

> You are an internal communications writer at a tech company. Rewrite the
> following policy text as a friendly, clear knowledge snippet for an employee
> newsletter. Requirements:
> - Use bullet points to organize the information
> - Paraphrase everything — do not quote the original directly
> - Keep the total output under 75 words
> - Only paraphrase the content in the input text; do not add new
>   recommendations or technologies
>
> Input text: "Employees must ensure that all remote access to internal
> systems is established via the approved secure VPN. Under no circumstances
> should unsecured connections or personal devices lacking endpoint protection
> be used to access proprietary data or sensitive communications."

**Example output:**

> **Working remotely? Keep it secure!**
> - Always connect to internal systems through the company-approved VPN
> - Never use unsecured networks to access company data
> - Personal devices without endpoint protection are off-limits for sensitive
>   files or communications
>
> A quick VPN check keeps our data safe — thanks for doing your part!

*(48 words)*

---

## Step 2: Evaluation

| Criteria | Assessment |
|---|---|
| Relevance | Sticks to VPN requirement, unsecured connections, personal devices |
| Clarity | Jargon like "proprietary data" simplified to "company data" |
| Structure | Clear bullet points with a short header |
| Tone | Friendly and encouraging, appropriate for internal comms |
| Length | 48 words, under the 75-word limit |
| Factual Accuracy | No policies invented; all three rules from the source preserved |

## Step 4: Paraphrase for a Junior Intern Audience

**Prompt:**

> Paraphrase the following policy for a junior intern who is new to the
> company. Rules:
> - Use plain, everyday language — no corporate or legal jargon
> - Maximum 4 short bullet points
> - Keep the tone supportive and informative
> - Only use information from the input text
>
> Input text: [policy paragraph]

**Example output:**

> - When working remotely, always use the company VPN to connect
> - Don't use public or unprotected Wi-Fi for work systems
> - Only use devices with company security software installed
> - This keeps company info (and you!) safe — ask IT if you're unsure

## Step 5: Quote Extraction Variant

**Prompt:**

> From the input text below, extract exactly one direct quote (verbatim, in
> quotation marks) that best captures the core security policy. Do not
> paraphrase or modify the wording.
>
> Input text: [policy paragraph]

**Example output:**

> "Employees must ensure that all remote access to internal systems is
> established via the approved secure VPN."

**When is quoting more appropriate than paraphrasing?**

In compliance-sensitive communications — policy acknowledgments, HR/legal
notices, audit documentation, or disciplinary contexts — where the exact
official wording matters and paraphrasing could change the legal meaning.

**When might quoting pose a risk?**

When the original text is dense or jargon-heavy, quoting can confuse
non-technical readers, get skimmed past, or be quoted out of context and
misinterpreted. It can also expose sensitive internal wording if shared
externally.
