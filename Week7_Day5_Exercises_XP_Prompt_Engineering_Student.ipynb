## Exercise 1: Rewrite and Optimize a Vague Prompt
**Scenario**: Manager asks for a LinkedIn post about FlowNest with the vague prompt `"Write something about productivity tips."`

### Tasks
- Identify 3 critical issues with the prompt.
- Rewrite it using: role prompting, clear instructions, and format/tone/length constraints.
- Add a final line to make it reusable for future social posts.

```python
# TODO: list three critical issues with the vague prompt
critical_issues = [
    'Lack of Context: It doesn\'t mention the product (FlowNest) or its unique value proposition.',
    'No Defined Audience: It doesn\'t specify who the tips are for (e.g., remote workers, students, managers).',
    'Missing Constraints: There is no guidance on tone, length, or platform-specific formatting (like LinkedIn hashtags or emojis).'
]
critical_issues
```

```python
# TODO: rewrite the prompt using role, instructions, and constraints
rewritten_prompt = """
Act as a Senior Social Media Manager for FlowNest, a productivity app.
Write a professional yet engaging LinkedIn post targeting busy remote professionals.

Instructions:
1. Share 3 actionable productivity tips that emphasize "deep work."
2. Naturally mention how FlowNest helps users stay in the flow state.
3. End with a strong call-to-action (CTA) to try the app.

Constraints:
- Tone: Encouraging and expert.
- Format: Use bullet points for tips.
- Length: Maximum 800 characters.
- Include 3 relevant hashtags.

Reusable line: Use this format whenever we promote productivity tools on social media by swapping the product name and specific tips.
"""
print(rewritten_prompt)
```

## Exercise 2: Multi-Part Prompt for Quiz Generation
**Scenario**: Create an educational prompt for 7th graders about volcanic eruptions.

### Tasks
- Ask for: 2-bullet summary + 3 MCQs (1 correct, 2 distractors).
- Require friendly tone, simple vocab (11?13 y/o), and Google Slides-ready output.
- Explain 3 ways your prompt improves on: `Make a quiz for kids about this article.`

```python
# TODO: write the full multi-part prompt
quiz_prompt = """
Act as an experienced Middle School Science Teacher.
Below is an article about volcanic eruptions. Please process it into educational content for 7th-grade students (11-13 years old).

Tasks:
1. Summarize the main points of the article into exactly 2 simple bullet points.
2. Create 3 multiple-choice questions (MCQs) based on the text. Each question must have:
   - One clear correct answer.
   - Two plausible but incorrect distractors.

Constraints:
- Vocabulary: Use simple language suitable for 11-13 year olds.
- Tone: Friendly and encouraging.
- Output Format: Organize the output into clear sections (Slide 1: Summary, Slide 2: Question 1, etc.) so it can be easily copied into Google Slides.

Article Text: [INSERT ARTICLE TEXT HERE]
"""
print(quiz_prompt)
```

```python
# TODO: explain three improvements over the vague prompt
improvements = [
    'Age-Appropriateness: By specifying the grade level (7th grade) and vocabulary constraints, we avoid overly technical jargon that the vague prompt might include.',
    'Structural Control: Defining the exact number of bullet points and the MCQ format (1 correct, 2 distractors) ensures the output is immediately usable without manual editing.',
    'Format for Purpose: Requesting a layout suitable for Google Slides prevents a "wall of text" and ensures the content is ready for the teacher\'s specific delivery method.'
]
print(improvements)
```

## Exercise 6: Hallucination Spotting and Mitigation
**Scenario**: Model added an unsupported claim in a climate/marine article summary.

### Tasks
1) Prompt the model to avoid hallucinations using only provided text.
2) Rewrite with verification language.
3) List 2 mitigation strategies.
4) Name 2 domains where hallucinations are risky and why.

```python
# TODO: baseline anti-hallucination prompt
no_hallucination_prompt = """
Summarize the provided article on climate change.

Constraint: Your summary must be based ONLY on the provided text. Do not incorporate outside knowledge or statistics not explicitly mentioned in the article.
"""
print(no_hallucination_prompt)
```

```python
# TODO: verification-focused rewrite
verification_prompt = """
Extract the key findings regarding marine biodiversity from the attached article.

Verification Rules:
1. Only include claims explicitly stated in the article.
2. If the article does not mention a specific statistic or projection, do not invent one.
3. If you are uncertain whether a piece of information is in the text, state: 'Information not available in the source.'
4. Provide the sentence from the text that supports each finding.
"""
print(verification_prompt)
```

```python
# TODO: list mitigation strategies and risky domains
mitigation_strategies = [
    'Grounded Generation (RAG mindset): Instructing the model to use only the provided context.',
    'Negative Constraints/Verification: Explicitly telling the model to admit when it doesn\'t know the answer instead of guessing.'
]
risky_domains = {
    'domain_1': {'name': 'Healthcare/Medical', 'why': 'Inaccurate medical advice or fake drug dosages can lead to physical harm or life-threatening situations.'},
    'domain_2': {'name': 'Legal', 'why': 'Hallucinated case law or statutes can lead to professional malpractice, unfair trials, or incorrect legal filings.'},
}
print(mitigation_strategies)
print(risky_domains)
```
