## Exercise 1: Open Source Levels Reflection
**Use:** Fully Open, Weights Released, Architecture Only; model components; fine-tuning/domain adaptation; healthcare compliance.

**Step-by-step**
1) Gather definitions for the three openness levels.  
2) Identify key characteristics (what is open; what you can/cannot do).  
3) Compare side by side in a table (what's open vs. impact).  
4) Write a 3?5 sentence comparative paragraph.  
5) Answer the healthcare prompt (1?2 sentences) about retraining on clinical data.

**Deliverables:** paragraph + healthcare answer.

```python
open_source_levels = {
    "Fully Open": {
        "definition": "Models where the source code, training data, and weights are all publicly available under a permissive license.",
        "what_is_open": ["Training Data", "Weights", "Pre-processing code"],
        "what_you_can_do": ["Audit data bias", "Modify architecture", "Full fine-tuning"],
        "what_you_cannot_do": ["Claim exclusive ownership", "Violate license attribution"],
    },
    "Weights Released": {
        "definition": "Models where the pre-trained weights are public, but the training data or specific code pipelines remain proprietary.",
        "what_is_open": ["Pre-trained weights", "Inference code"],
        "what_you_can_do": ["Fine-tune (LoRA/QLoRA)", "Self-host"],
        "what_you_cannot_do": ["View training data", "Retrain from scratch"],
    },
    "Architecture Only": {
        "definition": "Only the mathematical structure and layer configuration are shared, without the learned weights.",
        "what_is_open": ["Model structure", "Hyperparameters"],
        "what_you_can_do": ["Replicate architecture", "Train from scratch if you have data"],
        "what_you_cannot_do": ["Use the model immediately", "Access pre-learned knowledge"],
    },
}
display(open_source_levels)
```

```python
comparison_table = """| Openness level | What's open? | Impact on retraining/modifying |
| --- | --- | --- |
| Fully Open | Code + Data + Weights | Full control: allows deep auditing and data-level adjustments. |
| Weights Released | Model Weights | Practical control: allows efficient fine-tuning for specific tasks. |
| Architecture Only | Design/Blueprint | High cost: requires massive data/compute to become functional. |"""
print(comparison_table)
```

```python
comparative_paragraph = """L'ouverture des LLM varie du partage total (Fully Open) au simple partage de la structure (Architecture Only). Les modèles 'Fully Open' offrent la plus grande transparence pour l'audit de sécurité, tandis que les modèles 'Weights Released' comme Llama ou Mistral constituent le standard de l'industrie pour le déploiement rapide. Le niveau 'Architecture Only' est souvent réservé à la recherche académique ou aux acteurs disposant de ressources de calcul massives pour l'entraînement."""
print(comparative_paragraph)
```

```python
healthcare_prompt_answer = """Pour un assistant de santé utilisant des données cliniques sensibles, le niveau 'Fully Open' ou 'Weights Released' est essentiel. Cela permet un entraînement sur site (on-premise) garantissant la confidentialité des données et une adaptation précise au vocabulaire médical spécifique."""
print(healthcare_prompt_answer)
```
